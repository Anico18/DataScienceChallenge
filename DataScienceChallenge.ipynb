```python
# Importamos las librerías que utilizaremos para el tratamiento de datos
import pandas as pd
import numpy as np
import json
```

```python
# Cargamos los datos de la primera tabla
dropout = pd.read_csv('./dropout_90_dias_challenge.csv')
df_dropout = pd.DataFrame(dropout)
df_dropout
```

```python
# Cargamos la segunda tabla
empresa = pd.read_csv('./empresa_challenge.csv')
df_empresa = pd.DataFrame(empresa)
# Seleccionamos las columnas con mayor cantidad de características de valor.
df_empresa_premodel = df_empresa[['SEXO', 'ADH', 'PD', 'CSAT', 'AHT', 'email_identifier']].fillna(0)

    #df_empresa_filtered = df_empresa_model.query("ESTADO != 'ACTIVO'").copy()
    #subset_columns = ['email_identifier']
    #df_empresa_model = df_empresa_model.drop_duplicates(subset=subset_columns, keep='last')

# Seleccionamos y rellenamos los motivos de baja para brindar datos categóricos completos.
df_baja = df_empresa[['MOTIVO BAJA', 'ESTADO']].copy()
df_baja.loc[df_baja['ESTADO'] == 'ACTIVO', 'MOTIVO BAJA'] = df_baja.loc[df_baja['ESTADO'] == 'ACTIVO', 'MOTIVO BAJA'].fillna('CONTRATADO')
df_baja.loc[df_baja['ESTADO'] == 'INACTIVO', 'MOTIVO BAJA'] = df_baja.loc[df_baja['ESTADO'] == 'INACTIVO', 'MOTIVO BAJA'].fillna('NO ESPECIFICADO')
df_empresa_model = pd.concat([df_empresa_premodel, df_baja], axis="columns")

# Reemplazamos los valores de 'ACTIVO' e 'INACTIVO' por booleanos.
mapping = {'ACTIVO': 1, 'INACTIVO': 0}
df_empresa_model['ESTADO'] = df_empresa_model['ESTADO'].map(mapping)

# Convertimos el tipo de los valores de la columna 'email_identifier' a integer.
df_empresa_model['email_identifier'] = df_empresa_model['email_identifier'].astype(int)
df_empresa_model.to_csv('empresa_filtered.csv', index = False)
df_empresa_model
```

```python
# ------- OMITIDO ---------#

# Cargamos la tercera tabla y seleccionamos 'email_identifier'. Este paso se omitirá y se explicará en el vídeo de porqué se tomó esa decisión.
ocean = pd.read_csv('./ocean_challenge.csv')
df_ocean = pd.DataFrame(ocean)
df_ocean_model = df_ocean[['email_identifier']]
df_ocean_model
```

```python
# ------- OMITIDO ---------#

# Se selecciona y cargan los datos de la columna 'Meta' como json.
json_list = df_ocean['Meta'].apply(lambda x: json.loads(x))
json_list
```

```python
# ------- OMITIDO ---------#

# Se construye el data frame con los resultados de cada una de las encuestas por categoría.
labels = ["Extroversion", "Agreeableness", "Conscientiousness", "EmotionalStability", "OpennessToExperience"]
total_data = {label: [] for label in labels}

for record in json_list:
    score_data = record['score'] 
    for label in labels:
        total = score_data[label]['total']
        total_data[label].append(total)

df_meta_data = pd.DataFrame(total_data)
df_meta_data
```

```python
# ------- OMITIDO ---------#

# Combinamos con el 'email_identifier' para asignar los resultados a las personas con dicho identificador.
merged_ocean_df = df_ocean_model.join(df_meta_data, how='outer')
merged_ocean_df.to_csv('merged_ocean.csv', index = False)
merged_ocean_df
```

```python
# ------- OMITIDO ---------#

# Se unen los DataFrames de 'empresa' con 'ocean'. Los datos no coinciden en casi ninguna entrada.
merged_ocean_empresa = pd.merge(df_empresa_model, merged_ocean_df, on='email_identifier', how='outer')
merged_ocean_empresa.to_csv('merged_ocean_empresa.csv', index = False)
merged_ocean_empresa
```

```python
# Se combinan los DataFrames de 'empresa' con 'dropout' para saber quiénes se han retirado en 3 meses y bajo qué rendimiento.
merged_dropout_empresa = pd.merge(df_empresa_model, df_dropout, on='email_identifier', how='outer')
merged_dropout_empresa.to_csv('merged_dropout_empresa.csv', index = False)
merged_dropout_empresa
```

```python
# ------- OMITIDO ---------#

# Se combinan los 3 DataFrames. Existe demasiado NaN para optar por una técnica para trabajar esos datos.
merged_df = pd.merge(merged_dropout_empresa, merged_ocean_df, on='email_identifier', how='outer')
merged_df.to_csv('merged_df.csv', index=False)
merged_df
```

```python



# ETAPA DE MACHINE LEARNING


```

```python
#Importamos la data de SciKit Learn para entrenar un algoritmo.

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Realizamos el split de características y variable objetivo.

X = merged_dropout_empresa.drop('PERFORMANCE_GROUP', axis=1)  # Features
y = merged_dropout_empresa['PERFORMANCE_GROUP']  # Target variable
```

```python
# Definimos cuál es la información categórica que tenemos

categorical_features = X.select_dtypes(include=['object']).columns.tolist()
numerical_features = X.select_dtypes(exclude=['object']).columns.tolist()
```

```python
# Verificamos las características categóricas
categorical_features
```

```python
# Verificamos las características numéricas
numerical_features
```

```python
# Pre procesamos la información con escalado y encoding.
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_features),
    ('cat', OneHotEncoder(), categorical_features)
])
```

```python
# Definimos el clasificador como un Random Forest.

classifier = RandomForestClassifier(n_estimators=100, random_state=42)
```

```python
# Definimos el pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', classifier)
])
```

```python
# Realizamos el Train/Test Split con 30% de tamaño de test.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Entrenamos el modelo.
pipeline.fit(X_train, y_train)
```

```python
# Verificamos las predicciones.
y_pred = pipeline.predict(X_test)
y_pred
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("Classification report:")
print(classification_report(y_test, y_pred))
```
